##### Using Group-by’s and Merges
Group-by’s can be used to build groups of rows based on a specific feature in your dataset eg. the ‘Manufacturer’ categorical column. 

```python
group_by_Manufacturer = car_data.groupby(by=['Manufacturer'])

car_data_avg = round(group_by_Manufacturer.mean(),0)
car_data_avg
```

##### With the generated dataset above we are able to do better visualizations for the data set 

```python
# Since all the columns in car_data_count are the same, we will use just the first column as the rest yield the same result. iloc allows us to take all the rows and the zeroth column.

car_data_count_series = car_data_avg.iloc[5:5,1]


## we will assemble a DataFrame of only the relevant features to plot  
features_of_interest = pd.DataFrame({'EngineSize': car_data_avg['EngineSize'],
                                     'Passengers': car_data_avg['Passengers'],
                                   }
                                   )

## Let us plot a few of the manufactures for visibility
features_of_interest = features_of_interest.iloc[:10,]
features_of_interest.plot(kind='bar')




```

#### Want to try pie charts? Lets do it

##### Lets prepare some data. We can create a pie chart to show the population of categorical variables 

```python
# group by passengers 
group_by_Passengers = car_data.groupby(by=['Type'])

car_typecount = group_by_Passengers.count()

# get first column
car_typePassengers = pd.DataFrame(car_typecount.iloc[:,0])

car_typePassengers.rename(columns={'Manufacturer': 'Passengers'}, inplace=True)

car_typePassengers.plot.pie(subplots=True,figsize=(8, 8))

```

```python



# group by Type 
group_by_Type = car_data.groupby(by=['Type'])

car_typecount = group_by_Type.count()

car_typecount = pd.DataFrame(car_typecount.iloc[:,0])

car_typecount.rename(columns={'Manufacturer': 'Type'}, inplace=True)

car_typecount.plot.pie(subplots=True,figsize=(12, 8))
```

```python
# Group by Airbags
group_by_AirBags = car_data.groupby(by=['AirBags'])

car_AirBagscount = group_by_AirBags.count()

car_AirBagscount = pd.DataFrame(car_AirBagscount.iloc[:,0])

car_AirBagscount.rename(columns={'Manufacturer': 'Airbags'}, inplace=True)


car_AirBagscount.plot.pie(subplots=True,figsize=(5, 10))
```

```python
# Box plot for the revs per mile for: Audi, Hyundai, Suzuki, and Toyota
# Which of these manufacturers has the car with the highest revs per mile?

# Define manufactureres consisting required manufacturers
manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
selected_makers = car_data.loc[car_data['Manufacturer'].isin(manufacturers)] # use .loc to select row from the *indexed* column
# Plot box plot for the revs per mile
selected_makers.boxplot(by = 'Manufacturer', column = 'Rev.per.mile')
```

#### Answer to Q1: From the graph above, it seems Suzuki has the car with the highest revs per mile

```python
# Histogram of MPG in the city
# Histogram of MPG on the highway on the same axis
# Is it generally more fuel efficient to drive in the city or highway?

# Define relevant variables
mpg_city = car_data['MPG.city']
mpg_highway = car_data['MPG.highway']

# plot histogram containing MPG in the city and on the highway
plt.hist([mpg_city, mpg_highway], bins = 8, edgecolor = 'black',
         color = ['green', 'orange'], label = ['city', 'highway'])
#selected_makers.boxplot(by = 'Manufacturer', column = 'Rev.per.mile')

# add relevant labels
plt.title('Histogram of MPG')
plt.ylabel('MPG')
plt.legend(loc = 'upper right')
```

#### Answer to Question 2: The higher the MPG, the more economically friendly a car is. 
#### Basically, a high MPG means you can drive the same distance with less fuel in the tank.
#### Therefore, from the graph above, driving high way seems to have higher MPG, and more fuel efficient.

```python
# Lineplot showing the relationship between the Wheelbase and turning circle
# What is this relationship?
# What happens when the wheelbase gets larger?
# A lineplot does not seem suitable for given database
# A scatter plot is used instead to view relationship between the wheelbase and turning circle

wheelbase = car_data['Wheelbase']
turning_circle = car_data['Turn.circle']
plt.scatter(wheelbase, turning_circle)
plt.xlabel('Wheelbase')
plt.ylabel('Turning circle')
```

#### Answer to Question 3: From the graph above, relationship between the wheelbase and turning circle seem to be linear.
#### When the wheelbase gets larger, the turning circle also gets higher value

```python
# Bar plot showing the mean horsepower for each car type (small, midsize, etc.)
# Does a larger car mean more horsepower?
car_Type_Horsepower = car_data.groupby('Type')['Horsepower'].mean()
car_Type_Horsepower.plot(kind='bar')
plt.ylabel('Mean Horsepower')
plt.title('Mean horsepower for each car type')
```

#### Answer to Question 4: From the graph above, larger car generally mean more horsepower.

```python
# A pie chart for the respective engine sizes
# Which engine size seems to be the most popular?

# Group by enginer size and count
group_EngineSizecount = car_data.groupby('EngineSize').count()
group_EngineSizescount = pd.DataFrame(group_EngineSizecount.iloc[:, 0])
# Rename the count index colume to whitespace
group_EngineSizescount.rename(columns={"Unnamed: 0":""}, inplace=True)
 
# plot the pie chart
group_EngineSizescount.plot.pie(subplots=True, figsize=(8, 8))
plt.title('Engine Size Popularity')
```

#### Answer to Question 5 : The most popular engine size seems to be 3.0
